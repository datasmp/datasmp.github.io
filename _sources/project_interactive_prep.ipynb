```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

fig, ax = plt.subplots()

x = np.arange(0, 2*np.pi, 0.01)  
line, = ax.plot(x, np.sin(x))

def animate(i):
 line.set_ydata(np.sin(x + i/10.0))  # Update the data
 return line,

ani = animation.FuncAnimation(fig, animate, frames=60, interval=20, blit=True)

plt.close()
HTML(ani.to_html5_video() )
```

```python
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

fig, ax = plt.subplots()
ax.set_xlim(-3,3)
ax.set_ylim(0,9)
x = np.linspace(-3,3,100)
y = x**2

parabola, = ax.plot(-3, 9, marker='o', markersize=8, c='orange' )
ax.plot(x,y)

def animation_frame(i):
    parabola.set_xdata([i])    # sequence
    parabola.set_ydata([i**2])
    return parabola,

animation = FuncAnimation(fig, func = animation_frame, frames = np.arange(-3,3,0.01), interval = 3, blit=True)

plt.close()
HTML(animation.to_html5_video())
```
