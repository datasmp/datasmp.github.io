## Experimental Probability of Head

```python
N = 10

import random 
random.seed(0)

cases = ['H', 'T']
sample = []

for i in range(N):
    sample.append(cases[random.randint(0,1)])

sample.count('H'), sample.count('T')
```

```python
import matplotlib.pyplot as plt
import random 

random.seed(0)

cases = ['H', 'T']
N_list = range(3,10_001,10)
prob_H = []

for N in N_list:
    sample = []
    for i in range(N):
        sample.append(cases[random.randint(0,1)])
    prob_H.append(sample.count('H')/N)  

plt.figure(figsize=(20,4))
plt.plot(N_list, prob_H)
plt.hlines(0.5, 0, N, colors='red', linestyle='dashed')
plt.title('Experimental Probability of Head')
plt.ylabel('Prorbability of Head')
plt.xlabel('Number of flips')
plt.grid()
```
