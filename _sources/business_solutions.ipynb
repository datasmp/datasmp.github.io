## Revenue

### Question: Max of Revenue
Write a program which finds the $p$ which maximize the revenue for $0\le p \le 20$
- Plot the graph of revenue for $0\le p \le 20$.

```python
import matplotlib.pyplot as plt
p_values = range(1, 20)
revenue = [ p*(-1.5*p+30) for p in p_values ]
plt.plot(p_values, revenue)


max_revenue = 0
max_p = 0

for p in p_values:
    revenue = p*(-1.5*p+30) 
    if revenue > max_revenue:
        max_revenue = revenue
        max_p = p
        
print(f'Max Revenue: {max_revenue}  ---  p: {max_p}')
```

## Equilibrium Point
An equilibrium point is where the demand and supply curves intersect.

### Question: Equilibrium Point
For the given demand and the supply functions find the equilibrium point and plot it.

Demand: $p(q) = -0.03q + 1000$

Supply: $p(q) = 0.02q + 400$

```python
for q in range(20000):
    d = -0.03*q + 1000
    s = 0.02*q + 400
    if d == s:
        e1, e2 = q, d
        print(f'The equilibrium point is ({e1},{e2})')
        break
```

```python
q_values = range(0,20000)
demands = [-0.03*q + 1000 for q in q_values] 
supplies = [0.02*q + 400 for q in q_values] 

plt.plot(q_values, demands, label='demand', linestyle='dotted')
plt.plot(q_values, supplies, label='supply', linestyle='dotted')

plt.hlines(0,-1000,20000, color='k')
plt.vlines(0,-100,1200, color='k')
# plt.vlines(e1,0,e2, color='r')


plt.text(e1-2000, e2+250, f'equilibrium point ')
plt.text(e1-2000, e2+200, f'({e1},{e2})')
plt.arrow(e1, e2+180, 0,-100, width=70, head_length=30)

plt.legend();
```

## Linear Programming
Linear programming involves finding the highest or lowest possible outcome of a linear function, while satisfying specific constraints.

### Question: Linear Programming

The store sells three items labeled (a, b, c) priced at 2, 3, and 4 dollars respectively. The following conditions are known:

- The combined sales of items b and c exceed 90 units.
- The total sales of items a and b do not surpass 50 units.
- The sales of items a and c are more than 80 units.
- The store can sell up to 100 units of each item.

What is the highest revenue achievable, and at what quantities of items is this maximum revenue attained?
 

```python
max_value = 0
a_max, b_max, c_max = 0, 0, 0

for a in range(1, 100):
    for b in range(1,100):
        for c in range(1,100):
            if b+c > 90:
                if a+b<= 50:
                    if a+c > 80:
                        if 2*a+3*b+4*c > max_value:
                            max_value = 2*a+3*b+c
                            a_max, b_max, c_max = a,b,c

print(f'Max Value: {max_value}, a:{a_max},  b:{b_max},   c:{c_max}')
```

## Ordinary Annuity
An ordinary annuity is an interest earning account where you make equal payments at the end of each compounding period.

The future value of an ordinary annuity is given by the following formula.

$\displaystyle  A = R \left( \frac{(1+\frac{r}{m})^{mt}-1}{\frac{r}{m}}  \right)$

where,

- A: Future value
- R: regular payment
- r: annual interest rate
- m: number of payments in a year
- t: number of years

### Question: Ordinary Annuity
If you deposit 200 dollars every month into an account earning 5.25% interest per year, compounded monthly, how much money will you have saved after 30 years?

```python
R = 200
r = 0.0525
m = 12
t = 30

num = (1+r/m)**(m*t) -1 
den = r/m

A = R*num/den
A
```

## Amortized Loans
Amortization is paying off a loan by making equal payments.

The present value of an amortized loan is given by the following formula.

$\displaystyle  P = R \left( \frac{1-(1+\frac{r}{m})^{-mt}}{\frac{r}{m}}  \right)$

where,

- P: Present value
- R: regular payment
- r: annual interest rate
- m: number of payments in a year
- t: number of years

### Question: Amortized Loan
If you can make monthly payments of 350 dollars for 3 years at an annual interest rate of 9%, what is the present value of the loan you can take out?

```python
R = 350
r = 0.09
m = 12
t = 3

num = 1-(1+r/m)**(-m*t) 
den = r/m

P = R*(num/den)
P
```

### Question: Amortization schedule
If you take out a 449,000 dollars loan to buy a house with an annual interest rate of 2.75%, to be paid off over 30 years with a monthly payment of 1,833 dollars, construct an amortization schedule for the last 5 payments.

For each month, a fixed monthly payment is deducted from the balance. Additionally, simple interest is added to the loan based on that month's remaining balance.

```python
P = 449_000
r = 0.0275
m = 12
t = 30
R = 1833

balance = P
for i in range(1,m*t+1):
    simple_interest = balance * (r/12) * 1
    apply_principal = R - simple_interest
    balance -= apply_principal
    if i > m*t-5:
        print(f'Payment Number: {i}  Interest Paid: {simple_interest:.2f}  Paid on Principal: {apply_principal:.2f}  Balance: {balance:.2f}  ')
```
